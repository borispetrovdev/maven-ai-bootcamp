### Import Dependencies

```python
import pandas as pd
import openai

from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams, PointStruct
```

### Read the sampled dataset with Amazon inventory

### Preprocess title and features

```python
def preprocess_description(row):
  return f"{row['title']} {' '.join(row['features'])}"
```

```python
def extract_first_large_image(row):
  return row["images"][0].get("large", "")
```

```python
sampled_items["preprocessed_description"] = sampled_items.apply(preprocess_description, axis=1)
sampled_items["image"] = sampled_items.apply(extract_first_large_image, axis=1)
```

### Sample 50 items from the dataset

```python
df_sample = sampled_items.sample(n=50, random_state=42)
```

```python
len(df_sample)
```

```python
df_data_to_embed = df_sample[["preprocessed_description", "image", "rating_number", "price", "average_rating", "parent_asin"]]
```

```python
df_data_to_embed.head()
```

```python
if (not isinstance(df_data_to_embed, pd.DataFrame)):
    raise TypeError("expected a DataFrame")
data_to_embed = df_data_to_embed.to_dict(orient="records")
```

```python
data_to_embed
```

### Define the embedding function

```python
response = openai.embeddings.create(
    input="Random text",
    model="text-embedding-3-small"
)
```

```python
len(response.data[0].embedding)
```

```python
def get_embedding(text, model="text-embedding-3-small"):
  response = openai.embeddings.create(
      input=text,
      model=model
  )
  return response.data[0].embedding
```

```python
get_embedding("Hi")
```

### Create Qdrant Collection

```python
qdrant_client = QdrantClient(url="http://localhost:6333")
```

```python
qdrant_client.create_collection(
  collection_name="Amazon-items-collection-01",
  vectors_config=VectorParams(size=1536, distance=Distance.COSINE)
)
```

### Embed Data

#### Pointstruct example

```python
pointstruct = PointStruct(
  id=0,
  vector=get_embedding("Hello, world!"),
  payload={
    "text": "Hello, world!",
    "model": "text-embedding-3-small"
  }
)
```

```python
pointstruct
```

#### Amazon Data

```python
data_to_embed
```

```python
len(data_to_embed)
```

```python
pointstructs = []
for i, data in enumerate(data_to_embed):
  embedding = get_embedding(data["preprocessed_description"])
  pointstructs.append(
    PointStruct(
      id=i,
      vector=embedding,
      payload=data
    )
  ) 
```

```python
pointstructs
```

```python
len(pointstructs)
```

#### Write embedded data to Qdrant

```python
qdrant_client.upsert(
  collection_name="Amazon-items-collection-01",
  points=pointstructs
)
```

#### Define a function for data retrieval

```python
def retrieve_data(query, k=5):
  query_embedding = get_embedding(query)
  results = qdrant_client.query_points(
    collection_name="Amazon-items-collection-01",
    query=query_embedding,
    limit=k
  )
  return results
```

```python
retrieve_data("do you have anything that can help me attach my apple pencil to my ipad?", k=10).points
```
